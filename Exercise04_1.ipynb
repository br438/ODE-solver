```python
def is_even(x):
    # Returns True if the remainder of x / 2 is 0
    return x % 2 == 0
```

```python
## tests ##
assert is_even(0) == True
assert is_even(101) == False
assert is_even(982) == True
assert is_even(-5) == False
assert is_even(-8) == True

print("All tests passed!")
```
